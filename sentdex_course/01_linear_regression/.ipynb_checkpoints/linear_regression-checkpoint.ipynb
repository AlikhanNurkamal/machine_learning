```python
import pandas as pd
import numpy as np
import quandl
```

```python
df = quandl.get( "WIKI/GOOGL", authtoken = "48xD_o69QuVxKx-28vss" )
df.head( )
```

```python
# Columns high and low show the volatility of the stock prices during the day.
# Open and close columns depict if price has gone up or down in a day.
df = df[["Adj. Open", "Adj. High", "Adj. Low", "Adj. Close", "Adj. Volume"]]

# Let's create the column that depicts the high-low percentage.
df["HL_PCT"] = ( df["Adj. High"] - df["Adj. Low"] ) / df["Adj. Low"] * 100

# And another column that shows the trend of the price during the day (daily percent change).
df["change_PCT"] = ( df["Adj. Close"] - df["Adj. Open"] ) / df["Adj. Open"] * 100
```

```python
df.head( )
```

```python
# Now, I take only the necessary columns.
df = df[["Adj. Close", "HL_PCT", "change_PCT", "Adj. Volume"]]
df.head( )
```

### Here, "Adj. Close" column is believed to stand for the label (feature that I want to predict using other). However, it is not, because "HL_PCT" column directly depends on the "Adj. Close" column, so I cannot know the former until I predicted the latter and I cannot predict the latter unless I know the former feature.

```python
df.info( )
```

```python
forecast_col = "Adj. Close"
# df.fillna( -99999, inplace = True )

import math
forecast_out = int( math.ceil( 0.01 * len( df ) ) )

df["label"] = df[forecast_col].shift( -forecast_out )
df.head( )
```

```python
df.dropna( inplace = True )
df.tail( )
```

```python
from sklearn import preprocessing, svm
from sklearn.linear_model import LinearRegression
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.2 )
X_train = preprocessing.scale( X_train )
X_test = preprocessing.scale( X_test )
```

```python
lin_model = LinearRegression( )
lin_model.fit( X_train, y_train )
```

```python
accuracy = lin_model.score( X_test, y_test )
accuracy
```

```python
y_pred = lin_model.predict( X_test )
```

```python
lin_model_results = pd.DataFrame( { "True value" : y_test, "Predicted value" : y_pred } )
lin_model_results
```

```python
svm_model = svm.SVR( kernel = "linear" )
svm_model.fit( X_train, y_train )
```

```python
accuracy_svm = svm_model.score( X_test, y_test )
accuracy_svm
```

```python
y_pred_svm = svm_model.predict( X_test )
```

```python
svm_model_results = pd.DataFrame( { "True value" : y_test, "Predicted value" : y_pred_svm } )
svm_model_results
```

```python
import datetime
import matplotlib.pyplot as plt
%matplotlib inline

df["Forecast"] = np.nan
df
```

```python
last_date = df.iloc[-1].name
last_unix = last_date.timestamp( )
one_day = 86400
next_unix = last_unix + one_day
```

```python
for i in forecast_set:
    next_date = datetime.datetime.fromtimestamp( next_unix )
    next_unix += one_day
    df.loc[next_date] = [np.nan for _ in range( len( df.columns ) - 1 )] + [i]
    
plt.figure( figsize = ( 15, 9 ) )
df["Adj. Close"].plot( )
df["Forecast"].plot( )
plt.xlabel( "Date" )
plt.ylabel( "Stock price" )
plt.legend( )
plt.savefig( "Alphabet_stock.jpg", dpi = 300 )
plt.show( )
```
